```python
Matches_Played = pd.concat([
df["Team_1"],
df["Team_2"]
]).value_counts()
print(Matches_Played)
```

```python
matches_played_df = pd.concat([
    df[['Team_1']].rename(columns={'Team_1': 'Team'}),
    df[['Team_2']].rename(columns={'Team_2': 'Team'})
])
matches_played_counts = matches_played_df['Team'].value_counts()

wins = df['Winner'].value_counts()

win_df = pd.DataFrame({'Matches_Played': matches_played_counts, 'Wins': wins})
win_df['Win_Percentage'] = (win_df['Wins'] / win_df['Matches_Played']) * 100
win_df = win_df.fillna(0).sort_values(by='Win_Percentage', ascending=False)

print("Highest Win %:\n", win_df.head())
```

```python
df['Batting_First'] = df.apply(
    lambda row: row['Toss_winner'] if row['Toss_decision'] == 'Bat'
    else (row['Team_2'] if row['Toss_winner'] == row['Team_1'] else row['Team_1']),
    axis=1
)

df['Chasing_Win'] = df['Winner'] != df['Batting_First']
chasing_vs_defending = df['Chasing_Win'].value_counts().rename(index={True: 'Chasing Wins', False: 'Defending Wins'})

print("Chasing vs Defending Wins:\n", chasing_vs_defending)
```

```python
df['Score_of_first_innings'] = (
    df['Score_of_first_innings']
    .astype(str)
    .str.extract('(\d+)')  
    .astype(float)
)

# Now calculate average score by team
avg_score_by_team = (
    df.groupby('Team_1')['Score_of_first_innings']
    .mean()
    .sort_values(ascending=False)
)

print("Average 1st Innings Score by Team:\n", avg_score_by_team)
```

```python
all_teams = set(df['Team_1']).union(set(df['Team_2']))
teams_with_wins = set(df['Winner'].dropna().unique())
never_won_teams = all_teams - teams_with_wins

print("Teams that never won a match:", never_won_teams)
```

```python
df['Score_of_first_innings'] = (
    df['Score_of_first_innings']
    .astype(str)
    .str.extract('(\d+)')
    .astype(float)
)

plt.figure(figsize=(8,5))
sns.histplot(df['Score_of_first_innings'], bins=15, kde=True, color='skyblue')
plt.title("Distribution of First Innings Scores")
plt.xlabel("First Innings Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
df['Score_of_second_innings'] = (
    df['Score_of_second_innings']
    .astype(str)
    .str.extract('(\d+)')
    .astype(float)
)

plt.figure(figsize=(8,5))
sns.histplot(df['Score_of_second_innings'], bins=15, kde=True, color='orange')
plt.title("Distribution of Second Innings Scores")
plt.xlabel("Second Innings Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
avg_1st = df['Score_of_first_innings'].mean()
median_1st = df['Score_of_first_innings'].median()

print(f"Average First Innings Score: {avg_1st:.2f}")
print(f"Median First Innings Score: {median_1st:.2f}")
```

```python
df['Wkts_in_first_innings'] = pd.to_numeric(df['Wkts_in_first_innings'], errors='coerce')
df['Wkts_in_second_innings'] = pd.to_numeric(df['Wkts_in_second_innings'], errors='coerce')

avg_wkts_1st = df['Wkts_in_first_innings'].mean()
avg_wkts_2nd = df['Wkts_in_second_innings'].mean()

print(f"Avg Wickets in 1st Innings: {avg_wkts_1st:.2f}")
print(f"Avg Wickets in 2nd Innings: {avg_wkts_2nd:.2f}")
```

```python
df['Combined_Score'] = df['Score_of_first_innings'] + df['Score_of_second_innings']

max_combined = df.loc[df['Combined_Score'].idxmax(), [
    'Match_id', 'Team_1', 'Team_2', 'Score_of_first_innings', 'Score_of_second_innings', 'Combined_Score'
]]

print("Match with Highest Combined Score:\n", max_combined)
```

```python
toss_decision_counts = df['Toss_decision'].value_counts()
bat_pct = (toss_decision_counts.get('Bat', 0) / toss_decision_counts.sum()) * 100

print(f"% of teams that chose to bat after winning the toss: {bat_pct:.2f}%")
```

```python
toss_wins = df[df['Toss_winner'] == df['Winner']].shape[0]
total_matches = df['Match_id'].nunique()
toss_win_pct = (toss_wins / total_matches) * 100

print(f"Matches won by toss-winning team: {toss_wins} ({toss_win_pct:.2f}%)")
```

```python
def get_batting_first(row):
    if row['Toss_decision'] == 'Bat':
        return row['Toss_winner']
    else:
        return row['Team_2'] if row['Toss_winner'] == row['Team_1'] else row['Team_1']

df['Batting_First'] = df.apply(get_batting_first, axis=1)
df['Batting_Second'] = df.apply(
    lambda row: row['Team_1'] if row['Batting_First'] == row['Team_2'] else row['Team_2'], axis=1)

batting_first_wins = df[df['Winner'] == df['Batting_First']].shape[0]
batting_second_wins = df[df['Winner'] == df['Batting_Second']].shape[0]
total_decided = batting_first_wins + batting_second_wins

print(f"Batted First Win %: {(batting_first_wins / total_decided) * 100:.2f}%")
print(f"Chased Successfully %: {(batting_second_wins / total_decided) * 100:.2f}%")
```
