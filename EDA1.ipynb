```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Make sure the 'Cardiotocographic.csv' file is in the same directory as your notebook
df = pd.read_csv('Cardiotocographic.csv')

# Display the first few rows of the dataframe to get a sense of the data
print("Initial Data Preview:")
print(df.head())
print("\n")

# Check the data types and for any missing values
print("Initial DataFrame Info:")
df.info()
print("\n")
```

```python
# 2.1 Handle missing values
# Impute missing numerical values with the mean of the column.
# Note: For this assignment, we use mean imputation for simplicity. Other methods like median imputation or more advanced techniques can also be used.
df = df.apply(pd.to_numeric, errors='coerce')
df.fillna(df.mean(), inplace=True)
print("DataFrame after handling missing values:")
df.info()
print("\n")

# 2.2 Handle outliers (using IQR method)
# The assignment mentions detecting and treating outliers. We will use the IQR method
# to identify and cap outliers to the upper and lower bounds.
print("Treating Outliers with IQR Method...")
for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])

print("Outlier treatment complete. The data is now clean and ready for analysis.")
```

```python
# Provide a statistical summary for each variable
print("Statistical Summary of the Dataset:")
print(df.describe())
```

```python
# 4.1 Create Histograms for key numerical variables
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distribution of Key Variables', fontsize=20)

sns.histplot(df['LB'], bins=20, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Baseline Fetal Heart Rate (LB)')

sns.histplot(df['ASTV'], bins=20, kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Abnormal Short Term Variability (ASTV)')

sns.histplot(df['MSTV'], bins=20, kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Mean Short Term Variability (MSTV)')

sns.histplot(df['Width'], bins=20, kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Width of FHR Peaks (Width)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# 4.2 Create Boxplots to visualize distribution and outliers
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Boxplots of Select Variables', fontsize=20)

sns.boxplot(x=df['Width'], ax=axes[0], color='lightgray')
axes[0].set_title('Width of FHR Peaks')

sns.boxplot(x=df['ALTV'], ax=axes[1], color='lightgray')
axes[1].set_title('Abnormal Long Term Variability (ALTV)')

sns.boxplot(x=df['Tendency'], ax=axes[2], color='lightgray')
axes[2].set_title('Tendency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# 4.3 Bar chart for a categorical variable (NSP)
# The NSP column seems to be a categorical variable with values 1, 2, and 3.
# We will count the occurrences of each category.
# Updated code for the bar chart for the categorical variable (NSP)
plt.figure(figsize=(8, 6))
sns.countplot(x='NSP', data=df, palette='viridis', hue='NSP', legend=False)
plt.title('Frequency of Fetal State (NSP)', fontsize=16)
plt.xlabel('Fetal State (1=Normal, 2=Suspect, 3=Pathologic)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
# 4.4 Correlation Heatmap
# This helps to visualize the relationships between all pairs of variables.
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of All Variables', fontsize=16)
plt.show()
```

```python
# 4.5 Scatter plot for two variables
# Let's explore the relationship between Baseline FHR (LB) and Abnormal Short Term Variability (ASTV)
plt.figure(figsize=(10, 7))
sns.scatterplot(x='LB', y='ASTV', hue='NSP', data=df, palette='hsv')
plt.title('Scatter Plot of LB vs. ASTV by Fetal State', fontsize=16)
plt.xlabel('Baseline Fetal Heart Rate (LB)', fontsize=12)
plt.ylabel('Abnormal Short Term Variability (ASTV)', fontsize=12)
plt.show()
```

```python
# 4.6 Advanced Visualization: Pair Plot
# A pair plot is useful for showing the relationships between all pairs of variables
# in a single, comprehensive plot. This can provide deep insights into the dataset's structure.
# For demonstration, we'll plot a subset of columns.
print("Generating a Pair Plot for a subset of variables...")
columns_to_plot = ['LB', 'ASTV', 'MSTV', 'Width', 'NSP']
sns.pairplot(df[columns_to_plot], hue='NSP', palette='tab10')
plt.suptitle('Pair Plot of Key Variables', y=1.02, fontsize=16)
plt.show()
```
